```python
import os
import cv2
import numpy as np
import pathlib as pl
import matplotlib.pyplot as plt
print(os.getcwd())
```

```python
print(os.listdir(os.getcwd()))
path = 'images'
print(path)
files = os.listdir(path)
print(files)
```

```python
pl.Path('Result\\Reverse').mkdir(exist_ok=True)
for i, name in enumerate(files):
    print(str(i+1) + "ファイル目 : " + path + "\\" + name)
    img = cv2.imread(path + "\\" + name)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    print(img.shape)

    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            for z in range(img.shape[2]):
                img[x][y][z] = 255 - img[x][y][z]

    plt.figure(dpi=1000)
    plt.tick_params(labelbottom=False,
                    labelleft=False,
                    labelright=False,
                    labeltop=False,
                    bottom=False,
                    left=False,
                    right=False,
                    top=False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    plt.imshow(img)
    plt.grid(False)
    plt.savefig("Result\\reverse-" + name.split(".")[0] + ".png")
```

```python
pl.Path('Result\\Monotone').mkdir(exist_ok=True)
for i, name in enumerate(files):
    print(str(i+1) + "ファイル目 : " + path + "\\" + name)
    img = cv2.imread(path + "\\" + name)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    print(img.shape)

    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            temp = 0.3 * img[x][y][0] + 0.6 * img[x][y][1] + 0.1 * img[x][y][2]
            for z in range(img.shape[2]):
                img[x][y][z] = 255 - temp
                
    plt.figure(dpi=1000)
    plt.tick_params(labelbottom=False,
                    labelleft=False,
                    labelright=False,
                    labeltop=False,
                    bottom=False,
                    left=False,
                    right=False,
                    top=False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    plt.imshow(img)
    plt.grid(False)
    plt.savefig("Result\\Monotone\\mono-" + name.split(".")[0] + ".png")
```
