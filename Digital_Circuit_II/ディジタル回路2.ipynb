```python
import os
import numpy as np
import pandas as pd
import pathlib as pl
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.interpolate as ip
from matplotlib.font_manager import FontProperties as FP
%matplotlib inline
```

```python
font = FP(fname=r'C:\WINDOWS\Fonts\msgothic.ttc')
```

```python
data = pd.read_excel("data/ディジタル回路2.xlsx")
print(data)
```

```python
data_np = np.array(data)
volt = data_np[:,0]
curr = data_np[:,1]
print(data_np)
```

```python
plt.figure()
plt.xlabel("入力電圧[V]", fontproperties=font)
plt.ylabel("電源電流[mA]", fontproperties=font)
plt.plot(volt, curr, marker="o", label="電源電流")
plt.tick_params()
plt.grid(which="both")
plt.legend(prop=font)
```

```python
pl.Path('images\\Experiment').mkdir(exist_ok=True)
fig = plt.figure(dpi=300)
ax = fig.add_subplot(111)
ax.tick_params(direction='in', which='both', bottom=True, top=True, left=True, right=True)
ax.set_xlabel("入力電圧[V]", fontproperties=font, fontsize=14)
ax.set_xticks(np.linspace(0, 6, 31), minor=True)
ax.set_ylabel("電源電流[mA]", fontproperties=font, fontsize=14)
ax.set_yticks(np.linspace(0, 14, 36), minor=True)
ax.grid(which="major", color="lightGray")
ax.plot(volt, curr, marker="o", label="電源電流")
ax.legend(prop=font)
plt.savefig("images\\Experiment\\実験3_1.png")
```
